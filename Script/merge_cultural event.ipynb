```python
import pandas as pd
```

```python
df_meta = pd.read_csv('../Dataset/meta_data_date/2014-09-21.csv')
# df_meta.reset_index(True)
df_meta.drop(['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1'], axis=1, inplace=True)
df_meta.head(20)
```

```python
df_cultural = pd.read_csv('../Dataset/cultural_events_aarhus.csv', header=None, usecols=[12,16])
df_cultural.head(1)
```

```python
df_cultural.columns = ['timestamp', 'event']
df_cultural.head(5)
```

```python
df_cultural['date'] = df_cultural['timestamp'].apply(lambda x: x.split('T')[0])
df_cultural['time'] = df_cultural['timestamp'].apply(lambda x: x.split('T')[1].split(':')[0])
df_cultural.drop(['timestamp'], axis=1, inplace=True)
df_cultural
```

```python
df_cultural['time'] = pd.to_numeric(df_cultural['time'])
df_merge = pd.merge(df_meta, df_cultural, on=['date', 'time'], how='left')
df_merge.head(5)
```

```python

df_merge['event'].unique()
```

```python
import numpy as np
df_merge['event'] = df_merge['event'].replace(np.nan, 'No Event')
df_merge.head(5)
```
